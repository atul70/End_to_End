```python
df = df.replace("?", value=np.nan)
```

```python
df['Bore'] = df['Bore'].astype(float)
df['Stroke'] = df['Stroke'].astype(float)
df['normalized-losses'] = df['normalized-losses'].astype(float)
df['Peak-rpm'] = df['Peak-rpm'].astype(float)
df['Horsepower'] = df['Horsepower'].astype(float)
df['Price'] = df['Price'].astype(float)
```

```python
#Missing_Col = []
def finding_blank(data):
    Missing_Col = []
    for col in data.columns:
        count = data[col].isna().sum()
        print("Column Name is : ", col," <====And===> ", " Total null Value is : ", count)
        if count > 0:
            Missing_Col.append(col)
    print("===================================================")
    print(" \nMissing Data Column List==>\n  ", Missing_Col)
            
```

```python
finding_blank(df)
```

```python
df['Bore'].fillna((df['Bore'].mean()), inplace=True)
df['Stroke'].fillna((df['Stroke'].mean()), inplace=True)
df['normalized-losses'].fillna((df['normalized-losses'].mean()), inplace=True)
df['Peak-rpm'].fillna((df['Peak-rpm'].mean()), inplace=True)
df['Horsepower'].fillna((df['Horsepower'].mean()), inplace=True)
df['Price'].fillna((df['Price'].mean()), inplace=True)
df['Num-of-doors'].fillna((df['Num-of-doors'].mode()), inplace=True)
```

```python
df['normalized-losses'] = df['normalized-losses'].astype('int64')
df['Peak-rpm'] = df['Peak-rpm'].astype('int64')
df['Horsepower'] = df['Horsepower'].astype('int64')
df['Price'] = df['Price'].astype('int64')
```
