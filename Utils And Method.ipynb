# Data profiling

```python
class data_profiling:
    def __init__(self, data):
        pass
    
    def shape(self,data):
        col = []
        for i in data:
            col.append(i)
            row = len(data)
        print('Length of Row : ', row,'\n\n<==================>\n','\nLength of Columns : ', len(col))

    def question_marks(self,data):
        data = data.replace("?", value=np.nan)
        return data

    #total_count = 0    
    def missing_data(self,data):
        Missing_Col = []
        #total_count = 0
        for col in data.columns:
            total_count = 0
            count = data[col].isna().sum()
            percentage = (count/len(data))*100
            print("Column Name is : ", col," <====And===> ", " Total null Value is : ", count)
            print('Percentage of missing Value in {} is {} % : '.format(col, percentage))
            if count > 0:
                Missing_Col.append(col)
        total_count = count + total_count
        print("===================================================")
        print(" \nMissing Data Column List==>  ", Missing_Col)
        print('\n Total Missing Data is : ', total_count)
        print('\n Total percentage of missing data is {} % : '.format(total_count/len(data))*100)
        #return question_marks

    def catego_numerical(self,data):
        categorical_col = []
        numerical_col = []
        for col in data.columns:
            if data[col].dtypes == 'object':
                categorical_col.append(col)
            else:
                numerical_col.append(col)
        print('List of Categorical column : {} \n\nand their number is : {} '.format(categorical_col,len(categorical_col)))
        print('\nList of Numerical column : {} \n\nand their number is : {} '.format(numerical_col,len(numerical_col)))


    def duplicate_row(self,data):
        duplicate = data.duplicated().sum()
        print(duplicate)
        if duplicate > 0:
            duplicate.duplicated(keep = False)

    def zero_std_dvt(self,data):       
        return data.std()==0
    
    def memory_ram(self,data):
        process = psutil.Process(os.getpid())
        print(process.memory_info().rss,'bytes') 
```
